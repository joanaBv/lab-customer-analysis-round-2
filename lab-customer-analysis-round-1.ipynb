```python
# Column meanings and their optimal data types

# Index
# unique values to identify each row of the data frame
# "integer"

# Customer
# customer ID code
# in this case "object" as it includes letters and numbers 

# State
# state of residence for each customer
# "object"

# Customer Lifetime Value
# difference between the total amount of revenues from a customer and the companies' expenses for this customer during the whole duration of relationship​. 
# "integer"

# Education
# highest level of studies pursued by each customer
# "object"

# Gender
# gender of each customer
# "object"

# Income
# yearly income of each customer
# "integer"

# Monthly Premium Auto
# amount paid monthly by each customer to maintain the insurance policy
# "integer"

# Number of Open Complaints
# number of complaints currently active for each customer
# "integer"

# Policy Type
# type of policy offered
# "object"

# Total Claim Ammount
# the sum payable at the maturity of an insurance policy
# "integer"

# Vehicle Class
# type of car associated to each policy/customer
# "object"
```
