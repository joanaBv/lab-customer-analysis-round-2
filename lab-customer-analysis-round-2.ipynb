```python
# we've got several null values in the following categories:

# State
# Response
# Months since last claim
# Number of open complaints
# Vehicle Class
# Vehicle Size
# Vehicle Type

# for State, Response, Vehicle Class, Size and Type I will use the mode to
# fill in the null values
# so I need to see the value dispersion for each category

# As for the remaining columns, Months since last claim and Number of open
# complaints, I will use the mean value to fill in the null values
```

```python
# Start with "State"

marketing_customer_analysis_df['State'].value_counts(dropna=False)
```

```python
# For "State" the most common value is "California" by a considerable ammount,
# so I will use this answer to fill in the null values

marketing_customer_analysis_df['State'] = marketing_customer_analysis_df['State'].fillna('California')

# just checking

len(marketing_customer_analysis_df[marketing_customer_analysis_df['State'].isna()==True])
```

```python
# Same for "Response"

marketing_customer_analysis_df['Response'].value_counts(dropna=False)
```

```python
marketing_customer_analysis_df['Response'] = marketing_customer_analysis_df['Response'].fillna('No')

# just checking

len(marketing_customer_analysis_df[marketing_customer_analysis_df['Response'].isna()==True])
```

```python
# Same for "Vehicle Class"

marketing_customer_analysis_df['Vehicle Class'].value_counts(dropna=False)
```

```python
marketing_customer_analysis_df['Vehicle Class'] = marketing_customer_analysis_df['Vehicle Class'].fillna('Four-Door Car')

# just checking

len(marketing_customer_analysis_df[marketing_customer_analysis_df['Vehicle Class'].isna()==True])
```

```python
# Same for "Vehicle Size"

marketing_customer_analysis_df['Vehicle Size'].value_counts(dropna=False)
```

```python
marketing_customer_analysis_df['Vehicle Size'] = marketing_customer_analysis_df['Vehicle Size'].fillna('Medsize')

# just checking

len(marketing_customer_analysis_df[marketing_customer_analysis_df['Vehicle Size'].isna()==True])
```

```python
# Same for "Vehicle Type"

marketing_customer_analysis_df['Vehicle Type'].value_counts(dropna=False)
```

```python
# Seing as there isn't a trend here and most of the values here are null,
# I have opted to remove this column alltogether

marketing_customer_analysis_df = marketing_customer_analysis_df.drop(['Vehicle Type'], axis=1)
```

```python
# As for the remaining columns, Months since last claim and Number of open
# complaints, I will use the mean value to fill in the null values

import numpy as np

# first for "Months since last claim"

mean_months = marketing_customer_analysis_df['Months Since Last Claim'].mean()

marketing_customer_analysis_df['Months Since Last Claim'] = marketing_customer_analysis_df['Months Since Last Claim'].fillna(mean_months)

# then for "Number fo open complaints"

mean_complaints = marketing_customer_analysis_df['Number of Open Complaints'].mean()

marketing_customer_analysis_df['Number of Open Complaints'] = marketing_customer_analysis_df['Number of Open Complaints'].fillna(mean_complaints)

# just checking

print(len(marketing_customer_analysis_df[marketing_customer_analysis_df['Months Since Last Claim'].isna()==True]))

print(len(marketing_customer_analysis_df[marketing_customer_analysis_df['Number of Open Complaints'].isna()==True]))
```
